```python
# 导入所需的库
%matplotlib inline
from qiskit import QuantumCircuit,QuantumRegister,ClassicalRegister, execute, Aer, IBMQ
from qiskit.compiler import transpile, assemble
from qiskit.tools.jupyter import *
from qiskit.visualization import *
import numpy as np
```

```python
# 定义1比特的量子傅里叶变换
def qft_1():
    n=1
    circuit1 = QuantumCircuit(n,n)
    circuit1.h(0)
    return circuit1
```

```python
# 绘制电路
circuit1 = qft_1()
circuit1.measure(0, 0)
circuit1.draw(output='mpl')
```

```python
# 后端运行
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit1, simulator).result()
counts = result.get_counts(circuit1)
plot_histogram(counts, title='QFT counts')
```

```python
# 定义2比特的量子傅里叶变换
def qft_2():
    n=2
    circuit2 = QuantumCircuit(n,n)
    circuit2.h(1)
    circuit2.cu1(np.pi / 2, 0, 1)
    circuit2.h(0)
    circuit2.swap(0, 1)
    return circuit2
```

```python
# 绘制电路
circuit2 = qft_2()
circuit2.measure([0,1],[0,1])
circuit2.draw(output='mpl')
```

```python
# 后端运行
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit2, simulator).result()
counts = result.get_counts(circuit2)
plot_histogram(counts, title='QFT counts')
```
